```python
import numpy as np
from utils import split_data
```

```python
def mse(y_true,y_pred):
    n = len(y_true)
    return float((1/n) * sum(((y_true-y_pred)**2)))
```

```python
class LinearRegression:
    def __init__(self,n_iters=100,lr=0.1):
        self.weights = None
        self.bias = None
        self.n_iters= n_iters
        self.lr = lr
    def fit(self,X,y):
        n_samples, n_features = X.shape
        
        # initialize weights
        self.weights = np.random.rand(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            y_pred = np.dot(X,self.weights) + self.bias

            diff = y_pred - y

            # calculate the gradients of weights and bias
            dw = (1/n_samples) * np.dot(X.T,diff)
            db = (1/n_samples) * np.sum(diff)

            # update the weights and bias
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
    def predict(self,X):
        return np.dot(X,self.weights) + self.bias

        
```

```python
from sklearn.datasets import load_diabetes

diabetes_data = load_diabetes()

X  = diabetes_data['data']
y = diabetes_data['target']

X_train,X_test,y_train,y_test = split_data(X,y,test_size=0.1)
```

```python
regressor = LinearRegression(n_iters=5000)

regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_train)
val_loss = mse(y_train,y_pred)
print("Val loss: ",val_loss)
```
